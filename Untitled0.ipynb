```python
def image_augmentation(image, path):
  im1 =  cv2.flip(image,0)
  cv2.imwrite(path+'/cropped.jpg',im1)
  
  center = image.shape[1]/2, image.shape[0]/2
  angle = 30
  scale = 1.0
  center = (100, 100)
  im2 = cv2.getRotationMatrix2D(center, 0, 0.3)
  resized = cv2.warpAffine(image, im2, (image.shape[1], image.shape[0]))

  patch = (300, 200)
  center = (190, 220)
  im4 = cv2.getRectSubPix(image,patch,center)



  plt.subplot(2,3,1), plt.imshow(im1), plt.title("flipped")
  plt.subplot(2,3,2), plt.imshow(im2), plt.title("resize")
  plt.subplot(2,3,3), plt.imshow(im4), plt.title("cropping")
  plt.show()
```

```python
image_augmentation(image,"/content/augmente")
```
